```python
#Techniques for outlier detection and removal:
 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/nethajinirmal13/Training-datasets/main/diabetes.csv')
```

```python
df.sample(5)
```

```python
sns.distplot(df["BloodPressure"])
sns.distplot(df['BMI'])
```

# **TRIMMING**

```python
#Finding the Boundary Values
print("Highest allowed",df['BloodPressure'].mean() + 3*df['BloodPressure'].std())
print("Lowest allowed",df['BloodPressure'].mean() - 3*df['BloodPressure'].std())
```

```python
#Finding the Outliers
dfn=df[(df['BloodPressure'] > 127.172890) | (df['BloodPressure'] < 11.038047)]
dfn
```

```python
dfn.plot.hist()
```

```python
#Trimming of Outliers
new_df = df[(df['BloodPressure'] < 127.172890) & (df['BloodPressure'] > 11.038047)]
new_df
```

```python
sns.distplot(new_df["BloodPressure"])
#elemenated outliers
```

# CAPING TECHNIQUES FOR OUTLIERS


```python
#fix limits
upper_limit = df['BloodPressure'].mean() + 3*df['BloodPressure'].std()
lower_limit = df['BloodPressure'].mean() - 3*df['BloodPressure'].std()
```

```python
upper_limit,lower_limit
```

```python
#apply capping
df['BloodPressure'] = np.where(
    df['BloodPressure']>upper_limit,
    upper_limit,
    np.where(
        df['BloodPressure']<lower_limit,
        lower_limit,
        df['BloodPressure']
    )
)
```

```python
df['BloodPressure'].describe()
```

# IQR BASED Filtering

```python
sns.boxplot(df['Pregnancies'])
```

```python
#Finding the IQR
percentile25 = df['Pregnancies'].quantile(0.25)
percentile75 = df['Pregnancies'].quantile(0.75)
iqr=percentile75 - percentile25
print(iqr)
print(percentile25)
print(percentile75)
```

```python
#Finding upper and lower limit
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr
print(upper_limit)
print(lower_limit)
```

```python
#finding outliers
df[df['Pregnancies'] > upper_limit]
df[df['Pregnancies'] < lower_limit]
```

```python
#trimming
new_df = df[df['Pregnancies'] < upper_limit]
new_df.shape
```

```python
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(df['Pregnancies'])
plt.subplot(2,2,2)
sns.boxplot(df['Pregnancies'])
plt.subplot(2,2,3)
sns.distplot(new_df['Pregnancies'])
plt.subplot(2,2,4)
sns.boxplot(new_df['Pregnancies'])
plt.show()
```

```python
#capping
new_df_cap = df.copy()
new_df_cap['Pregnancies'] = np.where(
    new_df_cap['Pregnancies'] > upper_limit,
    upper_limit,
    np.where(
        new_df_cap['Pregnancies'] < lower_limit,
        lower_limit,
        new_df_cap['Pregnancies']
    )
)
```

```python
plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
sns.distplot(df['Pregnancies'])
plt.subplot(2,2,2)
sns.boxplot(df['Pregnancies'])
plt.subplot(2,2,3)
sns.distplot(new_df_cap['Pregnancies'])
plt.subplot(2,2,4)
sns.boxplot(new_df_cap['Pregnancies'])
plt.show()
```
